# 1 Data Partitioning
### train_test_split splis the data as you expect from the scikit learn library.

```python
def train_test_split(data, test_size, display=False):
    """
    data = io.loadmat("data/%s_data.mat" % data_name)
    data has fields: "test_data", "training_data", "training_labels"
    
    0 <= test_size <= 1
    
    modeled after train_test_split in sklearn.model_selection
    
    """
    assert 0 <= test_size <= 1
    # 1 extract fields we need before splitting
    X_train = data['training_data']
    y_train = data['training_labels']
    # 2 determining train and test number
    n = X_train.shape[0]
    train_num = int(n - test_size * n)
    # 3 shuffle the data
    shuffled_idx = np.array(range(n))
    np.random.shuffle(shuffled_idx)
    X_train = X_train[shuffled_idx]
    y_train = y_train[shuffled_idx]
    # 4 split the data
    X_train, X_test = X_train[:train_num], X_train[train_num:]
    y_train, y_test = y_train[:train_num], y_train[train_num:]
    return X_train, X_test, y_train, y_test
```

# Support Vector Machines: Coding
`visualize_training` is used to train model based on the number of sample specified, and display the output here.

```python
def visualize_training(data_name, samples, kernel='linear'):
    # loading data
    data = io.loadmat("data/%s_data.mat" % data_name)
    n, train, test = data['training_data'].shape[0], [], []
    model = svm.SVC(kernel=kernel)
    for count in samples:
        X_train, X_test, y_train, y_test = train_test_split(data, 1 - count / n)
        model.fit(X_train, y_train)
        train.append(model.score(X_train, y_train))
        test.append(model.score(X_test, y_test))
    plt.plot(samples, train, label='train')
    plt.plot(samples, test, label='val')
    plt.xlabel("Number of Training Samples")
    plt.ylabel("Accuracy")
    plt.title("Training Results for {}".format(data_name))
    plt.legend()
```

```python
%%time
visualize_training('mnist', [100, 200, 500, 1000, 2000, 5000, 10000], kernel='linear')
```

# Hyperparameter Tuning
`grid_search` is used to find the best set of hyperparameters possible for a given data_set, with `start`, `stop`, `num` indicating the lower and top bound for the linear search space. There is no option for log linear search.

```python
def grid_search(X_train, X_test, y_train, y_test, start, stop, num=10, printing=False, parallel=False, kernel='linear'):
    # set up score
    param_grid, test_scores = np.linspace(start, stop, num=num), []
    train_scores = []
    if printing:
        print("Number of training samples = {}".format(X_train.shape[0]))
    
    
    for i in range(len(param_grid)):
        param = param_grid[i]
        if kernel == 'linear':
            model = svm.LinearSVC(C=param).fit(X_train, y_train)
        else:
            model = svm.SVC(C=param, kernel=kernel).fit(X_train, y_train)
        train_scores.append(model.score(X_train, y_train))
        test_scores.append(model.score(X_test, y_test))
    if printing:
        plt.plot(param_grid, test_scores, label='test')
        plt.axhline(max(test_scores), c='red', label='max(scores)')
        plt.axvline(param_grid[np.argmax(test_scores)], c='green', 
                    label=str(param_grid[np.round(np.argmax(test_scores), 3)]))

        plt.xlabel("Hyperparameter Value for C")
        plt.ylabel("Accuracy")
        plt.title("Testing Results")
        plt.legend()
    
    return param_grid, test_scores
```

```python
%%time
data = io.loadmat("data/%s_data.mat" % 'mnist')
test_size = 0.80
param_grid, test_scores = grid_search(*train_test_split(data, test_size), 
                                      10**-8, 10**-6, num=5, printing=True, kernel='linear')
```

```python
best_c = param_grid[np.argmax(test_scores)]
print("best C is for mnist {}".format(best_c))
```

# K-Fold Cross Validation
The `k_fold_cv` function does hyperparameter search over a certain specified linear region while performing k-fold cross validation on it. Cross-validation is done by shuffling the indexes and then using them to re-order the training data.

```python
def k_fold_cv(data, start, stop, num=10, n_fold=5, printing=False, kernel='linear'):
    # 0 set up the data
    # 1 extract fields we need before splitting
    X = data['training_data']
    y = data['training_labels']
    # 2 determining train and test number
    n = X.shape[0]
    fold_size = n // n_fold
    # 3 shuffle the data
    shuffled_idx = np.array(range(n))
    np.random.shuffle(shuffled_idx)
    X = X[shuffled_idx]
    y = y[shuffled_idx]
    # 4 set up folds = [[f1], [f2], ..., [fn]]
    folds = []
    for i in range(n_fold - 1):
        folds.append(shuffled_idx[i * fold_size:(i + 1) * fold_size])
    folds.append(shuffled_idx[(n_fold - 1) * fold_size:])
    
    assert n == sum([len(fold) for fold in folds]), "folds must add up to total count"
    
    kfold_scores = [0 for _ in range(num)]
    
    for i in range(n_fold):
        print("Running Fold {}".format(i + 1))
        train_idx, test_idx = np.concatenate(folds[:i] + folds[i + 1:]), folds[i]
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        param_grid, test_scores = grid_search(X_train, X_test, y_train, y_test, start, stop, num=num, kernel=kernel)
        for i in range(num):
            kfold_scores[i] += test_scores[i]
    kfold_scores = [score / n_fold for score in kfold_scores]
    if printing:
        plt.plot(param_grid, kfold_scores, label='test')
        plt.axhline(max(kfold_scores), c='red', label='max(scores)')
        plt.axvline(param_grid[np.argmax(kfold_scores)], c='green', 
                    label=str(param_grid[np.round(np.argmax(kfold_scores), 3)]))

        plt.xlabel("Hyperparameter Value for C")
        plt.ylabel("Accuracy")
        plt.title("Testing Results")
        plt.legend()
    return param_grid, kfold_scores
```

```python
%%time
data = io.loadmat("data/%s_data.mat" % 'spam')
test_size = 0.70
param_grid, test_scores = grid_search(*train_test_split(data, test_size), 
                                      0.01, 500, num=200, kernel='linear', printing=True)
```

```python
%%time
data = io.loadmat("data/%s_data.mat" % 'spam')
param_grid, kfold_scores = k_fold_cv(data, 50, 56, num=100, n_fold=5, kernel='rbf', printing=True)
best_c = param_grid[np.argmax(kfold_scores)]
print("best C is for spam {}".format(best_c))
```
